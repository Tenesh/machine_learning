```python
x = [1,2,3,4]
y = [11,12,13,14]
```

```python
plt.plot(x,y);
```

```python
# 1st method
fig = plt.figure() # Creates a figure
ax = fig.add_subplot() # Add some axes
plt.show()
```

```python
# 2nd method
fig = plt.figure() # Creates a figure
ax = fig.add_subplot() # Add some axes 
ax.plot(x,y) # Add data 
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x,y);
```

## Making figure with NumPy arrays

```python
# Create some data
x = np.linspace(0, 10, 100)
x
```

```python
# Plot the data with line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data with scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Make plot from dictionary
nut_butter_price = {"Almond butter": 10,
                  "Peanut butter": 8}
fig, ax = plt.subplots()
ax.bar(nut_butter_price.keys(), nut_butter_price.values())
```

## Plot from Pandas dataframe

```python
# Make a dataframe
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2020', periods=1000))

ts = ts.cumsum()

ts.plot();
```

```python
car_sales["Price"] = car_sales["Price"].astype(str).str.replace('[\$\,\.]','').astype(int)
car_sales
```
